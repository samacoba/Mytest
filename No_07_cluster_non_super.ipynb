```python
#スーパースプレッダなし
import numpy as np

init_infe = 5 #初期の感染者数
i_prob = 0.2 #感染確率
cont_num = 10#接触者数

sample_num = 10000#10000回シミュレーション
epoch_num = 10#期間数(5日毎、10回)
```

```python
#シミュレーション
result = np.zeros((sample_num,epoch_num),dtype = np.int64)

for sample in range(sample_num):#シミュレーションのループ(10000回)
    
    infe = init_infe#初期の感染者数(5人)
    
    for epoch in range(epoch_num):#期間ループ(5日毎、10回)

        prob = np.random.rand(infe*cont_num)#現感染者×接触者分の乱数(0-1)を発生

        infe = (prob < i_prob).sum()#0.2未満となる乱数の個数カウントし、次の感染者数とする
        
        result[sample][epoch] =infe
```

```python
#結果テキスト出力（縦サンプル、横期間）
out =''
for sample in range(sample_num):    
    for epoch in range(epoch_num):
        out = out + str(result[sample][epoch]) +'\t'
    out = out +'\n' 
print(out)  
```

```python
#平均
ave = np.mean(result , axis=0)
for i in range(epoch_num):
    print(ave[i])
```

```python
#3回目（15日後）の状態
section = [1,21,51,101]

freq = np.zeros(5,dtype = np.int64)
res = result[:,2]

freq[0] = (res == 0).sum() 
for i in range(3):  
    freq[i+1] = ( (section[i] <= res) * (res < section[i+1])).sum()

freq[4] = ( section[3] <= res ).sum()

for i in range(5):
    print(freq[i])
```
